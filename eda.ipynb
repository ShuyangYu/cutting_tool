```python
import os
import gc
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

```python
path = {
    'trainSensorDataPath': 'data/01-TrainingData-qLua/02/Sensor/',
    'trainPlcDataPath': 'data/01-TrainingData-qLua/02/PLC/plc.csv',
    'testSensorDataPath': 'data/02-TestingData-keD1/03/Sensor/',
    'testPlcDataPath': 'data/02-TestingData-keD1/03/PLC/plc.csv'
}
```

## 训练集总行数

```python
def takeNumber(str):
    return int(str.split('/')[-1].split('.')[0])
```

```python
perTrainSensorPath = []
for dirpath, dirnames, filenames in os.walk(path['trainSensorDataPath']):
    for file in filenames:
        perTrainSensorPath.append(os.path.join(dirpath, file))
perTrainSensorPath.sort(key=takeNumber)
```

```python
plcTrain = pd.read_csv(path['trainPlcDataPath'])
```

```python
plcTrain.head()
```

```python
plcRows = plcTrain.groupby('csv_no').count().iloc[:, 0]
```

```python
for perPath, rows in zip(perTrainSensorPath, plcRows):
    sensor = pd.read_csv(perPath)
    _ = sensor.plot(x='current', figsize=(
        20, 12), title='Sensor plot for {}-{}'.format(perPath.split('/')[2], perPath.split('/')[-1]))
    plt.show()
    print('data : {:<7} rows : {:<14} plc rows : {}'.format(
        perPath.split('/')[-1], sensor.shape[0], rows))
```
